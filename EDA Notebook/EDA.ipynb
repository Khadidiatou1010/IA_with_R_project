```python
"""
==============================================================================
PARTIE 4 - ANALYSE DE LA QUALIT√â DES DONN√âES
==============================================================================
"""

# ==============================================================================
# 4. ANALYSE DE LA QUALIT√â DES DONN√âES
# ==============================================================================

print("\n" + "=" * 80)
print("4. ANALYSE DE LA QUALIT√â DES DONN√âES")
print("=" * 80 + "\n")

# 4.1 Analyse des valeurs manquantes
print("üîç 4.1 ANALYSE DES VALEURS MANQUANTES\n")
print("-" * 80)

# Calculer les statistiques des valeurs manquantes
missing_stats = pd.DataFrame({
    'Nombre': df.isnull().sum(),
    'Pourcentage': (df.isnull().sum() / len(df) * 100)
})
missing_stats = missing_stats.sort_values('Nombre', ascending=False)

# Afficher le tableau
print("Tableau r√©capitulatif des valeurs manquantes:\n")
print(missing_stats)
print()

# R√©sum√©
total_missing = df.isnull().sum().sum()
total_cells = df.shape[0] * df.shape[1]
missing_percentage = (total_missing / total_cells * 100)

print(f"üìä R√©sum√©:")
print(f"   ‚Ä¢ Total de valeurs manquantes: {total_missing}")
print(f"   ‚Ä¢ Total de cellules: {total_cells}")
print(f"   ‚Ä¢ Pourcentage global: {missing_percentage:.2f}%")
print()

if total_missing == 0:
    print("‚úì Aucune valeur manquante d√©tect√©e dans le dataset!")
else:
    print(f"‚ö†Ô∏è  {total_missing} valeurs manquantes d√©tect√©es")
print()

# Visualisation 1: Barplot horizontal des valeurs manquantes
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

if total_missing > 0:
    missing_percent = (df.isnull().sum() / len(df) * 100).sort_values(ascending=True)
    missing_percent = missing_percent[missing_percent > 0]

    # Barplot horizontal
    bars = axes[0].barh(missing_percent.index, missing_percent.values,
                        color=COLORS['warning'], edgecolor='black', alpha=0.8)
    axes[0].set_xlabel('Pourcentage de valeurs manquantes (%)', fontweight='bold')
    axes[0].set_ylabel('Variables', fontweight='bold')
    axes[0].set_title('üìä Pourcentage de Valeurs Manquantes par Variable',
                      fontsize=14, fontweight='bold', pad=20)
    axes[0].grid(axis='x', alpha=0.3)

    # Ajouter les valeurs sur les barres
    for i, (bar, val) in enumerate(zip(bars, missing_percent.values)):
        axes[0].text(val + max(missing_percent.values)*0.02, i, f'{val:.2f}%',
                     va='center', fontweight='bold')

    # Heatmap des valeurs manquantes
    missing_matrix = df.isnull().astype(int)
    if len(missing_matrix.columns) <= 20:  # Limiter si trop de colonnes
        sns.heatmap(missing_matrix.T, cmap='YlOrRd', cbar=True,
                    yticklabels=missing_matrix.columns, ax=axes[1],
                    cbar_kws={'label': 'Manquant (1) / Pr√©sent (0)'})
        axes[1].set_xlabel('Index des observations', fontweight='bold')
        axes[1].set_ylabel('Variables', fontweight='bold')
        axes[1].set_title('üî• Carte de Chaleur des Valeurs Manquantes',
                          fontsize=14, fontweight='bold', pad=20)
    else:
        # Si trop de colonnes, afficher un r√©sum√©
        axes[1].text(0.5, 0.5, f'Trop de variables pour\nafficher la heatmap\n({len(missing_matrix.columns)} colonnes)',
                    ha='center', va='center', fontsize=14, color='gray')
        axes[1].axis('off')
else:
    # Aucune valeur manquante
    for ax in axes:
        ax.text(0.5, 0.5, '‚úì Aucune valeur manquante',
                ha='center', va='center', fontsize=20, color='green', fontweight='bold')
        ax.axis('off')

plt.suptitle('üîç Analyse des Valeurs Manquantes', fontsize=18, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# 4.2 Analyse des doublons
print("\n" + "-" * 80)
print("üîç 4.2 ANALYSE DES DOUBLONS\n")

duplicates = df.duplicated().sum()
duplicate_percentage = (duplicates / len(df) * 100)

print(f"üìä R√©sum√©:")
print(f"   ‚Ä¢ Nombre de lignes dupliqu√©es: {duplicates}")
print(f"   ‚Ä¢ Pourcentage: {duplicate_percentage:.2f}%")

if duplicates == 0:
    print("   ‚úì Aucun doublon d√©tect√©!")
else:
    print(f"   ‚ö†Ô∏è  {duplicates} doublons d√©tect√©s")
    print("\n   Aper√ßu des lignes dupliqu√©es:")
    print(df[df.duplicated(keep=False)].sort_values(by=df.columns.tolist()).head(10))

print()

# Visualisation des doublons
fig, ax = plt.subplots(figsize=(8, 6))

if duplicates > 0:
    labels = ['Uniques', 'Doublons']
    sizes = [len(df) - duplicates, duplicates]
    colors_dup = [COLORS['success'], COLORS['warning']]
    explode = (0, 0.1)

    wedges, texts, autotexts = ax.pie(sizes, explode=explode, labels=labels,
                                        autopct='%1.1f%%', startangle=90,
                                        colors=colors_dup,
                                        textprops={'fontsize': 12, 'fontweight': 'bold'})
    for autotext in autotexts:
        autotext.set_color('white')
    ax.set_title('üîÑ Proportion de Doublons', fontsize=16, fontweight='bold', pad=20)
else:
    ax.text(0.5, 0.5, '‚úì Aucun doublon d√©tect√©',
            ha='center', va='center', fontsize=20, color='green', fontweight='bold')
    ax.axis('off')

plt.tight_layout()
plt.show()

# 4.3 Statistiques descriptives d√©taill√©es
print("\n" + "-" * 80)
print("üîç 4.3 STATISTIQUES DESCRIPTIVES\n")

# Variables num√©riques
if len(numerical_vars) > 0:
    print("üìà Statistiques descriptives - Variables Num√©riques:\n")
    numeric_stats = df[numerical_vars].describe().round(2)
    print(numeric_stats)
    print()

    # Statistiques suppl√©mentaires
    print("üìä Statistiques suppl√©mentaires (Asym√©trie et Aplatissement):\n")
    extra_stats = pd.DataFrame({
        'Skewness': df[numerical_vars].skew().round(3),
        'Kurtosis': df[numerical_vars].kurtosis().round(3),
        'Coefficient de Variation (%)': (df[numerical_vars].std() / df[numerical_vars].mean() * 100).round(2)
    })
    print(extra_stats)
    print()

# Variables cat√©gorielles
if len(categorical_vars) > 0:
    print("-" * 80)
    print("üìä Statistiques descriptives - Variables Cat√©gorielles:\n")

    cat_summary = []
    for var in categorical_vars:
        cat_summary.append({
            'Variable': var,
            'Nombre de cat√©gories': df[var].nunique(),
            'Cat√©gorie la plus fr√©quente': df[var].mode()[0] if len(df[var].mode()) > 0 else 'N/A',
            'Fr√©quence max': df[var].value_counts().iloc[0] if len(df[var]) > 0 else 0,
            'Pourcentage max (%)': (df[var].value_counts().iloc[0] / len(df) * 100).round(2) if len(df[var]) > 0 else 0
        })

    cat_df = pd.DataFrame(cat_summary)
    print(cat_df.to_string(index=False))
    print()

    # D√©tail de chaque variable cat√©gorielle
    print("\nüìã Distribution d√©taill√©e des variables cat√©gorielles:\n")
    for var in categorical_vars:
        print(f"   {var}:")
        value_counts = df[var].value_counts()
        value_percentages = (df[var].value_counts(normalize=True) * 100).round(2)

        for val, count, pct in zip(value_counts.index, value_counts.values, value_percentages.values):
            print(f"      ‚Ä¢ {val}: {count} ({pct}%)")
        print()

# 4.4 D√©tection des valeurs aberrantes (Outliers)
print("-" * 80)
print("üîç 4.4 D√âTECTION DES VALEURS ABERRANTES (OUTLIERS)\n")

if len(numerical_vars) > 0:
    outlier_summary = []

    for var in numerical_vars:
        Q1 = df[var].quantile(0.25)
        Q3 = df[var].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df[(df[var] < lower_bound) | (df[var] > upper_bound)][var]
        n_outliers = len(outliers)
        pct_outliers = (n_outliers / len(df) * 100)

        outlier_summary.append({
            'Variable': var,
            'Q1': Q1.round(2),
            'Q3': Q3.round(2),
            'IQR': IQR.round(2),
            'Limite inf√©rieure': lower_bound.round(2),
            'Limite sup√©rieure': upper_bound.round(2),
            'Nombre d\'outliers': n_outliers,
            'Pourcentage (%)': pct_outliers
        })

    outlier_df = pd.DataFrame(outlier_summary)
    print(outlier_df.to_string(index=False))
    print()

    # Visualisation des outliers avec boxplots
    n_num = len(numerical_vars)
    n_cols = min(3, n_num)
    n_rows = (n_num + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(6*n_cols, 5*n_rows))
    if n_num == 1:
        axes = [axes]
    else:
        axes = axes.ravel() if n_num > 1 else [axes]

    for idx, var in enumerate(numerical_vars):
        if idx < len(axes):
            bp = axes[idx].boxplot(df[var].dropna(), vert=True, patch_artist=True,
                                    boxprops=dict(facecolor=COLORS['primary'], alpha=0.7),
                                    medianprops=dict(color='red', linewidth=2),
                                    whiskerprops=dict(linewidth=1.5),
                                    capprops=dict(linewidth=1.5),
                                    flierprops=dict(marker='o', markerfacecolor=COLORS['warning'],
                                                   markersize=6, alpha=0.6))

            axes[idx].set_ylabel(var, fontweight='bold')
            axes[idx].set_title(f'Boxplot: {var}', fontsize=12, fontweight='bold')
            axes[idx].grid(axis='y', alpha=0.3)

            # Ajouter statistiques
            Q1 = df[var].quantile(0.25)
            Q3 = df[var].quantile(0.75)
            median = df[var].median()

            textstr = f'Q1: {Q1:.2f}\nM√©diane: {median:.2f}\nQ3: {Q3:.2f}'
            axes[idx].text(0.98, 0.98, textstr, transform=axes[idx].transAxes,
                          verticalalignment='top', horizontalalignment='right',
                          bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5),
                          fontsize=9)

    # Supprimer les axes vides
    for idx in range(len(numerical_vars), len(axes)):
        fig.delaxes(axes[idx])

    plt.suptitle('üì¶ D√©tection des Valeurs Aberrantes (M√©thode IQR)',
                 fontsize=18, fontweight='bold', y=1.00)
    plt.tight_layout()
    plt.show()

# 4.5 R√©sum√© de la qualit√©
print("\n" + "=" * 80)
print("üìã 4.5 R√âSUM√â DE LA QUALIT√â DES DONN√âES")
print("=" * 80 + "\n")

print(f"‚úì Dataset: {df.shape[0]} lignes √ó {df.shape[1]} colonnes")
print(f"‚úì Valeurs manquantes: {total_missing} ({missing_percentage:.2f}%)")
print(f"‚úì Doublons: {duplicates} ({duplicate_percentage:.2f}%)")

if len(numerical_vars) > 0:
    total_outliers = sum([row['Nombre d\'outliers'] for row in outlier_summary])
    print(f"‚úì Valeurs aberrantes d√©tect√©es: {total_outliers}")

print("\n‚ö†Ô∏è  RECOMMANDATIONS:")
if total_missing > 0:
    print("   ‚Ä¢ Traiter les valeurs manquantes (imputation ou suppression)")
if duplicates > 0:
    print("   ‚Ä¢ V√©rifier et traiter les doublons")
if len(numerical_vars) > 0 and total_outliers > 0:
    print("   ‚Ä¢ Analyser les outliers (conserver, transformer ou supprimer)")
if total_missing == 0 and duplicates == 0:
    print("   ‚úì Les donn√©es sont de bonne qualit√©!")

print("\n" + "=" * 80)
print("‚úì Section 4 termin√©e: Analyse de la qualit√© des donn√©es compl√®te")
print("=" * 80)
```

```python
"""
==============================================================================
PARTIE 6 - ANALYSE BIVARI√âE
==============================================================================
"""

# ==============================================================================
# 6. ANALYSE BIVARI√âE
# ==============================================================================

print("\n" + "=" * 80)
print("6. ANALYSE BIVARI√âE")
print("=" * 80 + "\n")

# 6.1 Boxplots de la variable cible par variables cat√©gorielles
print("üîç 6.1 RELATION ENTRE VARIABLES CAT√âGORIELLES ET VARIABLE CIBLE\n")
print("-" * 80 + "\n")

if len(categorical_vars) > 0 and target_var in df.columns:

    # 6.1.1 Boxplots classiques
    n_cat = len(categorical_vars)
    n_cols_cat = min(3, n_cat)
    n_rows_cat = (n_cat + n_cols_cat - 1) // n_cols_cat

    fig, axes = plt.subplots(n_rows_cat, n_cols_cat, figsize=(7*n_cols_cat, 6*n_rows_cat))

    if n_cat == 1:
        axes = [axes]
    else:
        axes = axes.ravel() if n_cat > 1 else [axes]

    for idx, cat_var in enumerate(categorical_vars):
        # Pr√©parer les donn√©es
        categories = df[cat_var].dropna().unique()
        data_by_cat = [df[df[cat_var] == cat][target_var].dropna() for cat in categories]

        # Boxplot
        bp = axes[idx].boxplot(data_by_cat, labels=categories, patch_artist=True,
                               boxprops=dict(facecolor=COLORS['primary'], alpha=0.7),
                               medianprops=dict(color='red', linewidth=2),
                               whiskerprops=dict(linewidth=1.5),
                               capprops=dict(linewidth=1.5),
                               flierprops=dict(marker='o', markerfacecolor=COLORS['warning'],
                                              markersize=6, alpha=0.6))

        axes[idx].set_xlabel(cat_var, fontweight='bold', fontsize=12)
        axes[idx].set_ylabel(f'{target_var} ($)', fontweight='bold', fontsize=12)
        axes[idx].set_title(f'{target_var} par {cat_var}', fontsize=14, fontweight='bold', pad=15)
        axes[idx].grid(axis='y', alpha=0.3, linestyle='--')

        # Rotation des labels si n√©cessaire
        if len(categories) > 3:
            axes[idx].set_xticklabels(categories, rotation=45, ha='right')

        # Ajouter les moyennes
        means = [data.mean() for data in data_by_cat]
        axes[idx].plot(range(1, len(means)+1), means, 'D', color='green',
                       markersize=8, label='Moyenne', zorder=5)
        axes[idx].legend(loc='upper right')

    # Supprimer les axes vides
    for idx in range(len(categorical_vars), len(axes)):
        fig.delaxes(axes[idx])

    plt.suptitle(f'üì¶ Boxplots: {target_var} par Variables Cat√©gorielles',
                 fontsize=18, fontweight='bold', y=1.00)
    plt.tight_layout()
    plt.show()

    # 6.1.2 Violin plots pour plus de d√©tails
    fig, axes = plt.subplots(n_rows_cat, n_cols_cat, figsize=(7*n_cols_cat, 6*n_rows_cat))

    if n_cat == 1:
        axes = [axes]
    else:
        axes = axes.ravel() if n_cat > 1 else [axes]

    for idx, cat_var in enumerate(categorical_vars):
        sns.violinplot(data=df, x=cat_var, y=target_var, ax=axes[idx],
                      palette="Set2", inner='box')

        axes[idx].set_xlabel(cat_var, fontweight='bold', fontsize=12)
        axes[idx].set_ylabel(f'{target_var} ($)', fontweight='bold', fontsize=12)
        axes[idx].set_title(f'Distribution de {target_var} par {cat_var}',
                           fontsize=14, fontweight='bold', pad=15)
        axes[idx].grid(axis='y', alpha=0.3, linestyle='--')

        # Rotation des labels si n√©cessaire
        categories = df[cat_var].unique()
        if len(categories) > 3:
            axes[idx].set_xticklabels(axes[idx].get_xticklabels(), rotation=45, ha='right')

    # Supprimer les axes vides
    for idx in range(len(categorical_vars), len(axes)):
        fig.delaxes(axes[idx])

    plt.suptitle(f'üéª Violin Plots: Distribution de {target_var}',
                 fontsize=18, fontweight='bold', y=1.00)
    plt.tight_layout()
    plt.show()

    # 6.1.3 Statistiques descriptives par cat√©gorie
    print("\nüìä Statistiques descriptives par cat√©gorie:\n")
    for cat_var in categorical_vars:
        print(f"\n{'='*60}")
        print(f"Variable: {cat_var}")
        print('='*60)

        grouped_stats = df.groupby(cat_var)[target_var].agg([
            ('Nombre', 'count'),
            ('Moyenne', 'mean'),
            ('M√©diane', 'median'),
            ('√âcart-type', 'std'),
            ('Min', 'min'),
            ('Q1', lambda x: x.quantile(0.25)),
            ('Q3', lambda x: x.quantile(0.75)),
            ('Max', 'max')
        ]).round(2)

        print(grouped_stats)
        print()

# 6.2 Tests statistiques (ANOVA/Kruskal-Wallis)
print("\n" + "-" * 80)
print("üîç 6.2 TESTS DE VARIANCE (Variable Cible vs Variables Cat√©gorielles)\n")
print("-" * 80 + "\n")

if len(categorical_vars) > 0 and target_var in df.columns:

    print("Tests utilis√©s:")
    print("  ‚Ä¢ ANOVA: si les donn√©es suivent une distribution normale")
    print("  ‚Ä¢ Kruskal-Wallis: si les donn√©es ne suivent pas une distribution normale\n")

    test_results = []

    for cat_var in enumerate(categorical_vars):
        if isinstance(cat_var, tuple):
            cat_var = cat_var[1]

        print(f"{'='*70}")
        print(f"Variable cat√©gorielle: {cat_var}")
        print('='*70)

        # Cr√©er les groupes
        groups = [df[df[cat_var] == cat][target_var].dropna()
                 for cat in df[cat_var].dropna().unique()]

        # Filtrer les groupes vides
        groups = [g for g in groups if len(g) > 0]

        if len(groups) < 2:
            print("‚ö†Ô∏è  Pas assez de groupes pour effectuer le test\n")
            continue

        # Test de normalit√© sur chaque groupe (Shapiro-Wilk)
        print(f"\n1Ô∏è‚É£  Test de normalit√© (Shapiro-Wilk) par groupe:")
        normal_groups = []
        for i, (cat, group) in enumerate(zip(df[cat_var].dropna().unique(), groups)):
            if len(group) >= 3:
                # Limiter √† 5000 observations pour Shapiro
                sample = group.sample(min(5000, len(group))) if len(group) > 5000 else group
                stat, p_val = stats.shapiro(sample)
                is_normal = p_val > 0.05
                normal_groups.append(is_normal)
                status = "‚úì Normale" if is_normal else "‚úó Non-normale"
                print(f"   {cat}: p-value = {p_val:.4f} {status}")
            else:
                normal_groups.append(False)
                print(f"   {cat}: Trop peu de donn√©es (n={len(group)})")

        # D√©cider quel test utiliser
        all_normal = all(normal_groups) if len(normal_groups) > 0 else False

        print(f"\n2Ô∏è‚É£  Test de variance:")

        if all_normal and len(groups) >= 2:
            # ANOVA
            f_stat, p_value = f_oneway(*groups)
            test_name = "ANOVA (F-test)"
            stat_name = "F-statistique"
        else:
            # Kruskal-Wallis
            f_stat, p_value = kruskal(*groups)
            test_name = "Kruskal-Wallis (H-test)"
            stat_name = "H-statistique"

        print(f"   Test utilis√©: {test_name}")
        print(f"   {stat_name}: {f_stat:.4f}")
        print(f"   P-value: {p_value:.4e}")

        # Interpr√©tation
        if p_value < 0.001:
            interpretation = "‚úì‚úì‚úì TR√àS SIGNIFICATIF (p < 0.001)"
        elif p_value < 0.01:
            interpretation = "‚úì‚úì TR√àS SIGNIFICATIF (p < 0.01)"
        elif p_value < 0.05:
            interpretation = "‚úì SIGNIFICATIF (p < 0.05)"
        else:
            interpretation = "‚úó NON SIGNIFICATIF (p ‚â• 0.05)"

        print(f"\n3Ô∏è‚É£  Interpr√©tation: {interpretation}")

        if p_value < 0.05:
            print(f"   ‚Üí Il existe une diff√©rence significative de {target_var}")
            print(f"     entre les cat√©gories de {cat_var}")
        else:
            print(f"   ‚Üí Pas de diff√©rence significative de {target_var}")
            print(f"     entre les cat√©gories de {cat_var}")

        print()

        # Stocker les r√©sultats
        test_results.append({
            'Variable': cat_var,
            'Test': test_name,
            'Statistique': f"{f_stat:.4f}",
            'P-value': f"{p_value:.4e}",
            'Significatif': "Oui" if p_value < 0.05 else "Non"
        })

    # Tableau r√©capitulatif
    if len(test_results) > 0:
        print("\n" + "="*70)
        print("TABLEAU R√âCAPITULATIF DES TESTS")
        print("="*70 + "\n")
        results_df = pd.DataFrame(test_results)
        print(results_df.to_string(index=False))
        print()

# 6.3 Relations entre variables num√©riques et variable cible
print("\n" + "-" * 80)
print("üîç 6.3 RELATION ENTRE VARIABLES NUM√âRIQUES ET VARIABLE CIBLE\n")
print("-" * 80 + "\n")

if len(numerical_features) > 0 and target_var in df.columns:

    # 6.3.1 Scatter plots avec r√©gression lin√©aire
    n_num = len(numerical_features)
    n_cols = min(3, n_num)
    n_rows = (n_num + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(7*n_cols, 6*n_rows))

    if n_num == 1:
        axes = [axes]
    else:
        axes = axes.ravel() if n_num > 1 else [axes]

    correlations = []

    for idx, num_var in enumerate(numerical_features):
        if idx < len(axes):
            # Donn√©es sans valeurs manquantes
            mask = df[num_var].notna() & df[target_var].notna()
            x_data = df.loc[mask, num_var]
            y_data = df.loc[mask, target_var]

            # Scatter plot
            axes[idx].scatter(x_data, y_data, alpha=0.5, color=COLORS['primary'],
                            s=30, edgecolors='black', linewidth=0.5)

            # R√©gression lin√©aire
            if len(x_data) > 1:
                z = np.polyfit(x_data, y_data, 1)
                p = np.poly1d(z)
                x_line = np.linspace(x_data.min(), x_data.max(), 100)
                axes[idx].plot(x_line, p(x_line), "r--", linewidth=2.5,
                             label='R√©gression lin√©aire')

            axes[idx].set_xlabel(num_var, fontweight='bold', fontsize=12)
            axes[idx].set_ylabel(f'{target_var} ($)', fontweight='bold', fontsize=12)
            axes[idx].set_title(f'{target_var} vs {num_var}', fontsize=14, fontweight='bold', pad=15)
            axes[idx].legend(loc='upper left')
            axes[idx].grid(alpha=0.3, linestyle='--')

            # Calculer corr√©lation
            if len(x_data) > 1:
                corr, p_val = pearsonr(x_data, y_data)
                correlations.append({
                    'Variable': num_var,
                    'Corr√©lation': corr,
                    'P-value': p_val
                })

                # Bo√Æte de texte avec corr√©lation
                textstr = f'r = {corr:.3f}\np = {p_val:.3e}'
                if p_val < 0.05:
                    textstr += '\n‚úì Significatif'
                    box_color = 'lightgreen'
                else:
                    textstr += '\n‚úó Non significatif'
                    box_color = 'lightcoral'

                axes[idx].text(0.05, 0.95, textstr,
                             transform=axes[idx].transAxes, verticalalignment='top',
                             bbox=dict(boxstyle='round', facecolor=box_color, alpha=0.7),
                             fontsize=10, fontweight='bold')

    # Supprimer axes vides
    for idx in range(len(numerical_features), len(axes)):
        fig.delaxes(axes[idx])

    plt.suptitle(f'üìà Relations: Variables Num√©riques vs {target_var}',
                 fontsize=18, fontweight='bold', y=1.00)
    plt.tight_layout()
    plt.show()

    # 6.3.2 Heatmap de corr√©lation avec le target
    if len(correlations) > 0:
        print("\nüìä Tableau des corr√©lations avec la variable cible:\n")
        corr_df = pd.DataFrame(correlations)
        corr_df['Corr√©lation'] = corr_df['Corr√©lation'].round(4)
        corr_df['P-value'] = corr_df['P-value'].apply(lambda x: f'{x:.4e}')
        corr_df['Force'] = corr_df['Corr√©lation'].apply(lambda x:
            'Tr√®s forte' if abs(x) > 0.8 else
            'Forte' if abs(x) > 0.6 else
            'Mod√©r√©e' if abs(x) > 0.4 else
            'Faible' if abs(x) > 0.2 else
            'Tr√®s faible'
        )
        corr_df['Direction'] = corr_df['Corr√©lation'].apply(lambda x:
            'Positive' if x > 0 else 'N√©gative'
        )

        print(corr_df.to_string(index=False))
        print()

    # 6.3.3 Visualisation en barres des corr√©lations
    if len(correlations) > 0:
        fig, ax = plt.subplots(figsize=(10, 6))

        vars_sorted = sorted(correlations, key=lambda x: abs(x['Corr√©lation']), reverse=True)
        variables = [v['Variable'] for v in vars_sorted]
        corr_values = [v['Corr√©lation'] for v in vars_sorted]
        colors_bars = [COLORS['success'] if c > 0 else COLORS['warning'] for c in corr_values]

        bars = ax.barh(variables, corr_values, color=colors_bars, edgecolor='black', alpha=0.8)
        ax.set_xlabel('Coefficient de Corr√©lation (Pearson)', fontweight='bold', fontsize=12)
        ax.set_ylabel('Variables', fontweight='bold', fontsize=12)
        ax.set_title(f'Corr√©lations avec {target_var}', fontsize=16, fontweight='bold', pad=20)
        ax.axvline(0, color='black', linewidth=1.5)
        ax.grid(axis='x', alpha=0.3, linestyle='--')

        # Ajouter les valeurs
        for bar, val in zip(bars, corr_values):
            width = bar.get_width()
            ax.text(width + (0.02 if width > 0 else -0.02), bar.get_y() + bar.get_height()/2,
                   f'{val:.3f}',
                   ha='left' if width > 0 else 'right', va='center', fontweight='bold')

        plt.tight_layout()
        plt.show()

# 6.4 Analyse crois√©e entre variables cat√©gorielles
print("\n" + "-" * 80)
print("üîç 6.4 ANALYSE CROIS√âE ENTRE VARIABLES CAT√âGORIELLES\n")
print("-" * 80 + "\n")

if len(categorical_vars) >= 2:

    # Tables de contingence et tests du Chi-2
    print("üìä Tables de contingence et tests d'ind√©pendance (Chi-2):\n")

    from itertools import combinations

    for var1, var2 in combinations(categorical_vars, 2):
        print(f"\n{'='*70}")
        print(f"{var1} vs {var2}")
        print('='*70)

        # Table de contingence
        contingency_table = pd.crosstab(df[var1], df[var2], margins=True)
        print("\nTable de contingence (effectifs):")
        print(contingency_table)

        # Table de contingence en pourcentages
        contingency_pct = pd.crosstab(df[var1], df[var2], normalize='all') * 100
        print("\nTable de contingence (pourcentages):")
        print(contingency_pct.round(2))

        # Test du Chi-2
        contingency_no_margins = pd.crosstab(df[var1], df[var2])
        chi2, p_value, dof, expected = chi2_contingency(contingency_no_margins)

        print(f"\nTest du Chi-2:")
        print(f"   Chi-2 statistique: {chi2:.4f}")
        print(f"   Degr√©s de libert√©: {dof}")
        print(f"   P-value: {p_value:.4e}")

        if p_value < 0.05:
            print(f"   ‚úì Les variables {var1} et {var2} sont D√âPENDANTES (p < 0.05)")
        else:
            print(f"   ‚úó Les variables {var1} et {var2} sont IND√âPENDANTES (p ‚â• 0.05)")
        print()

    # Heatmap des relations
    if len(categorical_vars) == 2:
        fig, ax = plt.subplots(figsize=(10, 6))
        contingency_pct = pd.crosstab(df[categorical_vars[0]], df[categorical_vars[1]],
                                      normalize='index') * 100

        sns.heatmap(contingency_pct, annot=True, fmt='.1f', cmap='YlOrRd',
                   cbar_kws={'label': 'Pourcentage (%)'}, ax=ax)
        ax.set_title(f'Heatmap: {categorical_vars[0]} vs {categorical_vars[1]}',
                    fontsize=16, fontweight='bold', pad=20)
        ax.set_xlabel(categorical_vars[1], fontweight='bold')
        ax.set_ylabel(categorical_vars[0], fontweight='bold')

        plt.tight_layout()
        plt.show()

else:
    print("‚ö†Ô∏è  Pas assez de variables cat√©gorielles pour l'analyse crois√©e.")

# R√©sum√©
print("\n" + "=" * 80)
print("‚úì Section 6 termin√©e: Analyse Bivari√©e compl√®te")
print("=" * 80)
```

```python
"""
==============================================================================
PARTIE 7 - ANALYSE DES CORR√âLATIONS
==============================================================================
"""

# ==============================================================================
# 7. ANALYSE DES CORR√âLATIONS
# ==============================================================================

print("\n" + "=" * 80)
print("7. ANALYSE DES CORR√âLATIONS")
print("=" * 80 + "\n")

# 7.1 Matrice de corr√©lation de Pearson
print("üîç 7.1 MATRICE DE CORR√âLATION (PEARSON)\n")
print("-" * 80 + "\n")

if len(numerical_vars) > 1:

    # Calculer la matrice de corr√©lation
    corr_matrix = df[numerical_vars].corr(method='pearson')

    print("üìä Matrice de corr√©lation compl√®te:\n")
    print(corr_matrix.round(3))
    print()

    # 7.1.1 Heatmap triangulaire (sans redondance)
    fig, axes = plt.subplots(1, 2, figsize=(18, 7))

    # Heatmap triangulaire
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

    sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.3f', cmap='coolwarm',
                center=0, square=True, linewidths=2, cbar_kws={"shrink": 0.8},
                vmin=-1, vmax=1, ax=axes[0],
                annot_kws={'fontsize': 10, 'fontweight': 'bold'})

    axes[0].set_title('üî• Matrice de Corr√©lation (Pearson) - Vue Triangulaire',
                      fontsize=14, fontweight='bold', pad=20)

    # Heatmap compl√®te avec bordures
    sns.heatmap(corr_matrix, annot=True, fmt='.3f', cmap='RdYlGn',
                center=0, square=True, linewidths=2, cbar_kws={"shrink": 0.8},
                vmin=-1, vmax=1, ax=axes[1],
                annot_kws={'fontsize': 10, 'fontweight': 'bold'})

    axes[1].set_title('üî• Matrice de Corr√©lation (Pearson) - Vue Compl√®te',
                      fontsize=14, fontweight='bold', pad=20)

    plt.suptitle('üìä Matrices de Corr√©lation', fontsize=18, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()

    # 7.1.2 Identifier les corr√©lations fortes
    print("\n" + "-" * 80)
    print("üîç Identification des corr√©lations significatives:\n")

    # Extraire les paires de corr√©lation (sans diagonale et sans doublons)
    correlation_pairs = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            var1 = corr_matrix.columns[i]
            var2 = corr_matrix.columns[j]
            corr_val = corr_matrix.iloc[i, j]

            correlation_pairs.append({
                'Variable 1': var1,
                'Variable 2': var2,
                'Corr√©lation': corr_val,
                'Corr√©lation (abs)': abs(corr_val)
            })

    # Trier par corr√©lation absolue
    correlation_pairs_df = pd.DataFrame(correlation_pairs)
    correlation_pairs_df = correlation_pairs_df.sort_values('Corr√©lation (abs)', ascending=False)

    # Classifier les corr√©lations
    def classify_correlation(corr):
        abs_corr = abs(corr)
        if abs_corr >= 0.8:
            return 'Tr√®s forte'
        elif abs_corr >= 0.6:
            return 'Forte'
        elif abs_corr >= 0.4:
            return 'Mod√©r√©e'
        elif abs_corr >= 0.2:
            return 'Faible'
        else:
            return 'Tr√®s faible'

    correlation_pairs_df['Force'] = correlation_pairs_df['Corr√©lation'].apply(classify_correlation)
    correlation_pairs_df['Direction'] = correlation_pairs_df['Corr√©lation'].apply(
        lambda x: 'Positive ‚Üó' if x > 0 else 'N√©gative ‚Üò'
    )

    # Afficher toutes les paires
    print("Toutes les paires de corr√©lation (tri√©es par force):\n")
    display_df = correlation_pairs_df[['Variable 1', 'Variable 2', 'Corr√©lation', 'Force', 'Direction']].copy()
    display_df['Corr√©lation'] = display_df['Corr√©lation'].round(4)
    print(display_df.to_string(index=False))
    print()

    # Corr√©lations fortes (|r| >= 0.6)
    strong_corr = correlation_pairs_df[correlation_pairs_df['Corr√©lation (abs)'] >= 0.6]
    if len(strong_corr) > 0:
        print("\n‚ö†Ô∏è  CORR√âLATIONS FORTES d√©tect√©es (|r| ‚â• 0.6):")
        for _, row in strong_corr.iterrows():
            print(f"   ‚Ä¢ {row['Variable 1']} ‚Üî {row['Variable 2']}: r = {row['Corr√©lation']:.4f} ({row['Force']})")
        print()

    # Corr√©lations mod√©r√©es (0.4 <= |r| < 0.6)
    moderate_corr = correlation_pairs_df[
        (correlation_pairs_df['Corr√©lation (abs)'] >= 0.4) &
        (correlation_pairs_df['Corr√©lation (abs)'] < 0.6)
    ]
    if len(moderate_corr) > 0:
        print("\nüìä Corr√©lations mod√©r√©es (0.4 ‚â§ |r| < 0.6):")
        for _, row in moderate_corr.iterrows():
            print(f"   ‚Ä¢ {row['Variable 1']} ‚Üî {row['Variable 2']}: r = {row['Corr√©lation']:.4f}")
        print()

    # Visualisation en r√©seau des corr√©lations fortes
    if len(strong_corr) > 0:
        fig, ax = plt.subplots(figsize=(12, 8))

        # Cr√©er un graphique en barres des corr√©lations fortes
        strong_pairs = [f"{row['Variable 1']} ‚Üî {row['Variable 2']}"
                       for _, row in strong_corr.iterrows()]
        strong_values = strong_corr['Corr√©lation'].values
        colors_strong = [COLORS['success'] if v > 0 else COLORS['warning'] for v in strong_values]

        bars = ax.barh(strong_pairs, strong_values, color=colors_strong,
                       edgecolor='black', alpha=0.8)
        ax.set_xlabel('Coefficient de Corr√©lation', fontweight='bold', fontsize=12)
        ax.set_title('üìä Corr√©lations Fortes entre Variables (|r| ‚â• 0.6)',
                    fontsize=16, fontweight='bold', pad=20)
        ax.axvline(0, color='black', linewidth=1.5)
        ax.grid(axis='x', alpha=0.3, linestyle='--')

        # Ajouter les valeurs
        for bar, val in zip(bars, strong_values):
            width = bar.get_width()
            ax.text(width + (0.02 if width > 0 else -0.02),
                   bar.get_y() + bar.get_height()/2,
                   f'{val:.3f}',
                   ha='left' if width > 0 else 'right', va='center',
                   fontweight='bold', fontsize=11)

        plt.tight_layout()
        plt.show()

else:
    print("‚ö†Ô∏è  Pas assez de variables num√©riques pour calculer une matrice de corr√©lation.")

# 7.2 Tests de significativit√© des corr√©lations
print("\n" + "-" * 80)
print("üîç 7.2 TESTS DE SIGNIFICATIVIT√â DES CORR√âLATIONS\n")
print("-" * 80 + "\n")

if len(numerical_vars) > 1:

    print("Test de significativit√©: Coefficient de Pearson avec p-value\n")
    print("H‚ÇÄ: La corr√©lation est nulle (r = 0)")
    print("H‚ÇÅ: La corr√©lation est diff√©rente de z√©ro (r ‚â† 0)\n")

    # Matrice de p-values
    p_values_matrix = pd.DataFrame(np.zeros((len(numerical_vars), len(numerical_vars))),
                                   columns=numerical_vars, index=numerical_vars)

    # Calculer les p-values pour chaque paire
    for i, var1 in enumerate(numerical_vars):
        for j, var2 in enumerate(numerical_vars):
            if i != j:
                mask = df[var1].notna() & df[var2].notna()
                if mask.sum() > 2:
                    _, p_val = pearsonr(df.loc[mask, var1], df.loc[mask, var2])
                    p_values_matrix.iloc[i, j] = p_val
                else:
                    p_values_matrix.iloc[i, j] = 1.0
            else:
                p_values_matrix.iloc[i, j] = 0.0

    print("üìä Matrice des p-values:\n")
    print(p_values_matrix.round(4))
    print()

    # Heatmap des p-values
    fig, axes = plt.subplots(1, 2, figsize=(18, 7))

    # Heatmap de significativit√© (p-value < 0.05)
    significance_matrix = (p_values_matrix < 0.05).astype(int)
    mask_tri = np.triu(np.ones_like(significance_matrix, dtype=bool))

    sns.heatmap(significance_matrix, mask=mask_tri, annot=True, fmt='d',
                cmap='RdYlGn', center=0.5, square=True, linewidths=2,
                cbar_kws={"shrink": 0.8, "ticks": [0, 1],
                         "label": "0=Non significatif, 1=Significatif"},
                vmin=0, vmax=1, ax=axes[0],
                annot_kws={'fontsize': 11, 'fontweight': 'bold'})

    axes[0].set_title('‚úì Matrice de Significativit√© (p < 0.05)',
                      fontsize=14, fontweight='bold', pad=20)

    # Heatmap des p-values
    sns.heatmap(p_values_matrix, mask=mask_tri, annot=True, fmt='.4f',
                cmap='RdYlGn_r', square=True, linewidths=2,
                cbar_kws={"shrink": 0.8, "label": "P-value"},
                vmin=0, vmax=0.1, ax=axes[1],
                annot_kws={'fontsize': 9})

    axes[1].set_title('üìä Matrice des P-values',
                      fontsize=14, fontweight='bold', pad=20)

    plt.suptitle('üî¨ Tests de Significativit√© des Corr√©lations',
                 fontsize=18, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()

    # Tableau d√©taill√© avec corr√©lations et p-values
    print("\n" + "-" * 80)
    print("üìã Tableau d√©taill√©: Corr√©lations avec p-values\n")

    detailed_corr = []
    for i in range(len(numerical_vars)):
        for j in range(i+1, len(numerical_vars)):
            var1 = numerical_vars[i]
            var2 = numerical_vars[j]

            mask = df[var1].notna() & df[var2].notna()
            if mask.sum() > 2:
                corr_val, p_val = pearsonr(df.loc[mask, var1], df.loc[mask, var2])

                detailed_corr.append({
                    'Variable 1': var1,
                    'Variable 2': var2,
                    'Corr√©lation (r)': f"{corr_val:.4f}",
                    'P-value': f"{p_val:.4e}",
                    'Significatif (Œ±=0.05)': '‚úì Oui' if p_val < 0.05 else '‚úó Non',
                    'Significatif (Œ±=0.01)': '‚úì Oui' if p_val < 0.01 else '‚úó Non',
                })

    detailed_df = pd.DataFrame(detailed_corr)
    print(detailed_df.to_string(index=False))
    print()

# 7.3 Corr√©lations avec la variable cible
print("\n" + "-" * 80)
print(f"üîç 7.3 CORR√âLATIONS AVEC LA VARIABLE CIBLE ({target_var})\n")
print("-" * 80 + "\n")

if target_var in numerical_vars and len(numerical_features) > 0:

    print(f"Analyse d√©taill√©e des corr√©lations avec {target_var}:\n")

    target_correlations = []

    for var in numerical_features:
        mask = df[var].notna() & df[target_var].notna()
        if mask.sum() > 2:
            # Pearson
            corr_pearson, p_pearson = pearsonr(df.loc[mask, var], df.loc[mask, target_var])

            # Spearman (pour d√©tecter relations non-lin√©aires)
            corr_spearman, p_spearman = spearmanr(df.loc[mask, var], df.loc[mask, target_var])

            target_correlations.append({
                'Variable': var,
                'Pearson (r)': corr_pearson,
                'P-value Pearson': p_pearson,
                'Spearman (œÅ)': corr_spearman,
                'P-value Spearman': p_spearman
            })

    # Cr√©er le DataFrame
    target_corr_df = pd.DataFrame(target_correlations)
    target_corr_df = target_corr_df.sort_values('Pearson (r)', key=abs, ascending=False)

    # Ajouter classifications
    target_corr_df['Force (Pearson)'] = target_corr_df['Pearson (r)'].apply(classify_correlation)
    target_corr_df['Significatif'] = target_corr_df['P-value Pearson'].apply(
        lambda x: '‚úì‚úì‚úì' if x < 0.001 else '‚úì‚úì' if x < 0.01 else '‚úì' if x < 0.05 else '‚úó'
    )

    # Afficher le tableau
    display_target = target_corr_df.copy()
    display_target['Pearson (r)'] = display_target['Pearson (r)'].round(4)
    display_target['P-value Pearson'] = display_target['P-value Pearson'].apply(lambda x: f'{x:.4e}')
    display_target['Spearman (œÅ)'] = display_target['Spearman (œÅ)'].round(4)
    display_target['P-value Spearman'] = display_target['P-value Spearman'].apply(lambda x: f'{x:.4e}')

    print(display_target.to_string(index=False))
    print()

    # Interpr√©tation
    print("\nüí° INTERPR√âTATION:\n")

    for _, row in target_corr_df.iterrows():
        var = row['Variable']
        r = row['Pearson (r)']
        p = row['P-value Pearson']
        force = row['Force (Pearson)']

        direction = "positive" if r > 0 else "n√©gative"

        if p < 0.05:
            print(f"   ‚Ä¢ {var}:")
            print(f"     - Corr√©lation {force.lower()} et {direction} (r = {r:.4f})")
            print(f"     - Relation statistiquement significative (p = {p:.4e})")

            if abs(r) > 0.6:
                print(f"     ‚ö†Ô∏è  IMPORTANT: Forte corr√©lation d√©tect√©e!")
        else:
            print(f"   ‚Ä¢ {var}: Pas de corr√©lation significative (p = {p:.4e})")
        print()

    # Visualisation comparative Pearson vs Spearman
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Pearson
    vars_sorted = target_corr_df.sort_values('Pearson (r)')['Variable'].tolist()
    pearson_vals = target_corr_df.set_index('Variable').loc[vars_sorted, 'Pearson (r)'].values
    colors_p = [COLORS['success'] if v > 0 else COLORS['warning'] for v in pearson_vals]

    bars_p = axes[0].barh(vars_sorted, pearson_vals, color=colors_p,
                          edgecolor='black', alpha=0.8)
    axes[0].set_xlabel('Coefficient de Corr√©lation (Pearson)', fontweight='bold', fontsize=12)
    axes[0].set_title(f'Corr√©lation de Pearson avec {target_var}',
                      fontsize=14, fontweight='bold', pad=15)
    axes[0].axvline(0, color='black', linewidth=1.5)
    axes[0].grid(axis='x', alpha=0.3, linestyle='--')

    for bar, val in zip(bars_p, pearson_vals):
        width = bar.get_width()
        axes[0].text(width + (0.02 if width > 0 else -0.02),
                    bar.get_y() + bar.get_height()/2,
                    f'{val:.3f}',
                    ha='left' if width > 0 else 'right', va='center',
                    fontweight='bold', fontsize=10)

    # Spearman
    spearman_vals = target_corr_df.set_index('Variable').loc[vars_sorted, 'Spearman (œÅ)'].values
    colors_s = [COLORS['success'] if v > 0 else COLORS['warning'] for v in spearman_vals]

    bars_s = axes[1].barh(vars_sorted, spearman_vals, color=colors_s,
                          edgecolor='black', alpha=0.8)
    axes[1].set_xlabel('Coefficient de Corr√©lation (Spearman)', fontweight='bold', fontsize=12)
    axes[1].set_title(f'Corr√©lation de Spearman avec {target_var}',
                      fontsize=14, fontweight='bold', pad=15)
    axes[1].axvline(0, color='black', linewidth=1.5)
    axes[1].grid(axis='x', alpha=0.3, linestyle='--')

    for bar, val in zip(bars_s, spearman_vals):
        width = bar.get_width()
        axes[1].text(width + (0.02 if width > 0 else -0.02),
                    bar.get_y() + bar.get_height()/2,
                    f'{val:.3f}',
                    ha='left' if width > 0 else 'right', va='center',
                    fontweight='bold', fontsize=10)

    plt.suptitle(f'üìä Comparaison Pearson vs Spearman avec {target_var}',
                 fontsize=18, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()

    # Scatter plot comparatif
    if len(numerical_features) >= 2:
        print("\n" + "-" * 80)
        print("üìà Visualisation: Top 2 variables corr√©l√©es avec la cible\n")

        top_2_vars = target_corr_df.nlargest(2, 'Pearson (r)', keep='first')['Variable'].tolist()

        fig, axes = plt.subplots(1, 2, figsize=(16, 6))

        for idx, var in enumerate(top_2_vars[:2]):
            mask = df[var].notna() & df[target_var].notna()
            x_data = df.loc[mask, var]
            y_data = df.loc[mask, target_var]

            axes[idx].scatter(x_data, y_data, alpha=0.5, color=COLORS['primary'],
                            s=40, edgecolors='black', linewidth=0.5)

            # R√©gression
            z = np.polyfit(x_data, y_data, 1)
            p = np.poly1d(z)
            x_line = np.linspace(x_data.min(), x_data.max(), 100)
            axes[idx].plot(x_line, p(x_line), "r--", linewidth=3,
                          label='R√©gression lin√©aire')

            axes[idx].set_xlabel(var, fontweight='bold', fontsize=12)
            axes[idx].set_ylabel(f'{target_var} ($)', fontweight='bold', fontsize=12)
            axes[idx].set_title(f'{target_var} vs {var}', fontsize=14, fontweight='bold', pad=15)
            axes[idx].legend()
            axes[idx].grid(alpha=0.3, linestyle='--')

            # Stats
            corr_val = target_corr_df[target_corr_df['Variable'] == var]['Pearson (r)'].values[0]
            p_val = target_corr_df[target_corr_df['Variable'] == var]['P-value Pearson'].values[0]

            textstr = f'r = {corr_val:.3f}\np = {p_val:.3e}'
            axes[idx].text(0.05, 0.95, textstr, transform=axes[idx].transAxes,
                          verticalalignment='top',
                          bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.7),
                          fontsize=11, fontweight='bold')

        plt.suptitle(f'üìà Top 2 Corr√©lations avec {target_var}',
                     fontsize=18, fontweight='bold', y=1.02)
        plt.tight_layout()
        plt.show()

# R√©sum√©
print("\n" + "=" * 80)
print("‚úì Section 7 termin√©e: Analyse des Corr√©lations compl√®te")
print("=" * 80)
```

```python
"""
==============================================================================
PARTIE 8 - ANALYSE MULTIVARI√âE
==============================================================================
"""

# ==============================================================================
# 8. ANALYSE MULTIVARI√âE
# ==============================================================================

print("\n" + "=" * 80)
print("8. ANALYSE MULTIVARI√âE")
print("=" * 80 + "\n")

# 8.1 Pairplot des variables num√©riques
print("üîç 8.1 PAIRPLOT DES VARIABLES NUM√âRIQUES\n")
print("-" * 80 + "\n")

if len(numerical_vars) >= 2:

    print("Cr√©ation du pairplot avec toutes les variables num√©riques...")
    print("(Cela peut prendre quelques secondes)\n")

    # Cr√©er une copie des donn√©es pour le pairplot
    pairplot_data = df[numerical_vars].copy()

    # Cr√©er des cat√©gories de la variable cible pour la coloration
    if target_var in numerical_vars:
        # Diviser en quartiles
        pairplot_data[f'{target_var}_cat'] = pd.qcut(
            pairplot_data[target_var],
            q=4,
            labels=['Q1 (Faible)', 'Q2 (Moyen-Bas)', 'Q3 (Moyen-Haut)', 'Q4 (√âlev√©)'],
            duplicates='drop'
        )

        # Pairplot avec coloration par quartile
        pairplot = sns.pairplot(
            pairplot_data,
            hue=f'{target_var}_cat',
            palette='viridis',
            diag_kind='kde',
            plot_kws={'alpha': 0.6, 's': 30, 'edgecolor': 'black', 'linewidth': 0.3},
            diag_kws={'alpha': 0.7, 'linewidth': 2}
        )

        pairplot.fig.suptitle(f'üîÄ Pairplot des Variables Num√©riques (color√© par quartiles de {target_var})',
                             fontsize=16, fontweight='bold', y=1.01)

    else:
        # Pairplot simple sans coloration
        pairplot = sns.pairplot(
            pairplot_data,
            diag_kind='kde',
            plot_kws={'alpha': 0.6, 's': 30, 'edgecolor': 'black', 'linewidth': 0.3},
            diag_kws={'alpha': 0.7, 'linewidth': 2}
        )

        pairplot.fig.suptitle('üîÄ Pairplot des Variables Num√©riques',
                             fontsize=16, fontweight='bold', y=1.01)

    plt.tight_layout()
    plt.show()

    print("‚úì Pairplot g√©n√©r√© avec succ√®s!\n")
    print("üí° Le pairplot permet de visualiser:")
    print("   ‚Ä¢ Les distributions univari√©es (diagonale)")
    print("   ‚Ä¢ Les relations bivari√©es (hors diagonale)")
    print("   ‚Ä¢ Les patterns et clusters potentiels")
    print()

else:
    print("‚ö†Ô∏è  Pas assez de variables num√©riques pour cr√©er un pairplot.\n")

# 8.2 Analyse par variables cat√©gorielles
print("\n" + "-" * 80)
print("üîç 8.2 ANALYSE MULTIVARI√âE PAR VARIABLES CAT√âGORIELLES\n")
print("-" * 80 + "\n")

if len(categorical_vars) >= 1 and len(numerical_vars) >= 2:

    # Choisir la variable cat√©gorielle avec le moins de cat√©gories pour la visualisation
    cat_var_for_analysis = min(categorical_vars, key=lambda x: df[x].nunique())

    print(f"Analyse d√©taill√©e par: {cat_var_for_analysis}\n")
    print(f"Nombre de cat√©gories: {df[cat_var_for_analysis].nunique()}\n")

    # 8.2.1 Statistiques descriptives group√©es
    print("üìä Statistiques descriptives par cat√©gorie:\n")

    grouped_stats = df.groupby(cat_var_for_analysis)[numerical_vars].agg(['mean', 'median', 'std'])
    print(grouped_stats.round(2))
    print()

    # 8.2.2 Boxplots multiples
    n_num_vars = len(numerical_vars)
    n_cols = min(3, n_num_vars)
    n_rows = (n_num_vars + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(7*n_cols, 5*n_rows))

    if n_num_vars == 1:
        axes = [axes]
    else:
        axes = axes.ravel() if n_num_vars > 1 else [axes]

    for idx, num_var in enumerate(numerical_vars):
        if idx < len(axes):
            df.boxplot(column=num_var, by=cat_var_for_analysis, ax=axes[idx],
                      patch_artist=True, grid=False)
            axes[idx].set_xlabel(cat_var_for_analysis, fontweight='bold', fontsize=11)
            axes[idx].set_ylabel(num_var, fontweight='bold', fontsize=11)
            axes[idx].set_title(f'{num_var} par {cat_var_for_analysis}',
                              fontsize=13, fontweight='bold', pad=15)
            axes[idx].get_figure().suptitle('')
            axes[idx].grid(axis='y', alpha=0.3, linestyle='--')

    # Supprimer axes vides
    for idx in range(n_num_vars, len(axes)):
        fig.delaxes(axes[idx])

    plt.suptitle(f'üì¶ Distribution des Variables Num√©riques par {cat_var_for_analysis}',
                 fontsize=18, fontweight='bold', y=1.00)
    plt.tight_layout()
    plt.show()

    # 8.2.3 Violin plots group√©s
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(7*n_cols, 5*n_rows))

    if n_num_vars == 1:
        axes = [axes]
    else:
        axes = axes.ravel() if n_num_vars > 1 else [axes]

    for idx, num_var in enumerate(numerical_vars):
        if idx < len(axes):
            sns.violinplot(data=df, x=cat_var_for_analysis, y=num_var,
                          ax=axes[idx], palette='Set2', inner='box')
            axes[idx].set_xlabel(cat_var_for_analysis, fontweight='bold', fontsize=11)
            axes[idx].set_ylabel(num_var, fontweight='bold', fontsize=11)
            axes[idx].set_title(f'{num_var} par {cat_var_for_analysis}',
                              fontsize=13, fontweight='bold', pad=15)
            axes[idx].grid(axis='y', alpha=0.3, linestyle='--')

            # Rotation si n√©cessaire
            if df[cat_var_for_analysis].nunique() > 3:
                axes[idx].set_xticklabels(axes[idx].get_xticklabels(),
                                         rotation=45, ha='right')

    # Supprimer axes vides
    for idx in range(n_num_vars, len(axes)):
        fig.delaxes(axes[idx])

    plt.suptitle(f'üéª Violin Plots par {cat_var_for_analysis}',
                 fontsize=18, fontweight='bold', y=1.00)
    plt.tight_layout()
    plt.show()

# 8.3 Analyse par r√©gion (si la variable existe)
if 'region' in df.columns:
    print("\n" + "-" * 80)
    print("üîç 8.3 ANALYSE D√âTAILL√âE PAR R√âGION\n")
    print("-" * 80 + "\n")

    print("üìä Vue d'ensemble par r√©gion:\n")

    # 8.3.1 Statistiques par r√©gion
    region_summary = df.groupby('region').agg({
        target_var: ['count', 'mean', 'median', 'std', 'min', 'max']
    }).round(2)

    print(region_summary)
    print()

    # 8.3.2 Visualisations multiples par r√©gion
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))

    # Graphique 1: Charges moyennes par r√©gion
    region_charges = df.groupby('region')[target_var].mean().sort_values(ascending=False)
    bars1 = axes[0, 0].bar(range(len(region_charges)), region_charges.values,
                           color=COLORS['secondary'], edgecolor='black', alpha=0.8)
    axes[0, 0].set_xticks(range(len(region_charges)))
    axes[0, 0].set_xticklabels(region_charges.index, rotation=45, ha='right')
    axes[0, 0].set_xlabel('R√©gion', fontweight='bold', fontsize=11)
    axes[0, 0].set_ylabel(f'{target_var} Moyen ($)', fontweight='bold', fontsize=11)
    axes[0, 0].set_title(f'{target_var} Moyen par R√©gion',
                        fontsize=13, fontweight='bold', pad=15)
    axes[0, 0].grid(axis='y', alpha=0.3, linestyle='--')

    for i, (bar, v) in enumerate(zip(bars1, region_charges.values)):
        height = bar.get_height()
        axes[0, 0].text(bar.get_x() + bar.get_width()/2., height,
                       f'${v:,.0f}',
                       ha='center', va='bottom', fontweight='bold', fontsize=10)

    # Graphique 2: Distribution de l'√¢ge par r√©gion
    if 'age' in df.columns:
        df.boxplot(column='age', by='region', ax=axes[0, 1], patch_artist=True, grid=False)
        axes[0, 1].set_xlabel('R√©gion', fontweight='bold', fontsize=11)
        axes[0, 1].set_ylabel('√Çge', fontweight='bold', fontsize=11)
        axes[0, 1].set_title('Distribution de l\'√Çge par R√©gion',
                            fontsize=13, fontweight='bold', pad=15)
        axes[0, 1].get_figure().suptitle('')
        axes[0, 1].grid(axis='y', alpha=0.3, linestyle='--')

    # Graphique 3: BMI par r√©gion
    if 'bmi' in df.columns:
        df.boxplot(column='bmi', by='region', ax=axes[1, 0], patch_artist=True, grid=False)
        axes[1, 0].set_xlabel('R√©gion', fontweight='bold', fontsize=11)
        axes[1, 0].set_ylabel('BMI', fontweight='bold', fontsize=11)
        axes[1, 0].set_title('Distribution du BMI par R√©gion',
                            fontsize=13, fontweight='bold', pad=15)
        axes[1, 0].get_figure().suptitle('')
        axes[1, 0].grid(axis='y', alpha=0.3, linestyle='--')

    # Graphique 4: Proportion de fumeurs par r√©gion
    if 'smoker' in df.columns:
        smoker_region = pd.crosstab(df['region'], df['smoker'], normalize='index') * 100
        smoker_region.plot(kind='bar', ax=axes[1, 1],
                          color=[COLORS['success'], COLORS['warning']],
                          edgecolor='black', alpha=0.8)
        axes[1, 1].set_xlabel('R√©gion', fontweight='bold', fontsize=11)
        axes[1, 1].set_ylabel('Pourcentage (%)', fontweight='bold', fontsize=11)
        axes[1, 1].set_title('Proportion de Fumeurs par R√©gion',
                            fontsize=13, fontweight='bold', pad=15)
        axes[1, 1].legend(title='Fumeur', labels=['Non', 'Oui'])
        axes[1, 1].grid(axis='y', alpha=0.3, linestyle='--')
        axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45, ha='right')

    plt.suptitle('üó∫Ô∏è Analyse Multivari√©e par R√©gion',
                 fontsize=18, fontweight='bold', y=0.995)
    plt.tight_layout()
    plt.show()

# 8.4 Analyse crois√©e : Variable cible par combinaisons de variables cat√©gorielles
print("\n" + "-" * 80)
print("üîç 8.4 ANALYSE CROIS√âE : VARIABLE CIBLE PAR COMBINAISONS\n")
print("-" * 80 + "\n")

if len(categorical_vars) >= 2 and target_var in df.columns:

    # Prendre les 2 premi√®res variables cat√©gorielles
    cat1, cat2 = categorical_vars[0], categorical_vars[1]

    print(f"Analyse de {target_var} par {cat1} et {cat2}\n")

    # 8.4.1 Tableau crois√© des moyennes
    print(f"üìä Tableau crois√© des moyennes de {target_var}:\n")
    pivot_mean = df.pivot_table(values=target_var, index=cat1, columns=cat2, aggfunc='mean')
    print(pivot_mean.round(2))
    print()

    # 8.4.2 Heatmap du tableau crois√©
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Heatmap des moyennes
    sns.heatmap(pivot_mean, annot=True, fmt='.0f', cmap='YlOrRd',
                cbar_kws={'label': f'{target_var} Moyen ($)'}, ax=axes[0],
                linewidths=2, linecolor='white')
    axes[0].set_title(f'Heatmap: Moyenne de {target_var} par {cat1} et {cat2}',
                     fontsize=14, fontweight='bold', pad=15)
    axes[0].set_xlabel(cat2, fontweight='bold')
    axes[0].set_ylabel(cat1, fontweight='bold')

    # Heatmap des effectifs
    pivot_count = df.pivot_table(values=target_var, index=cat1, columns=cat2, aggfunc='count')
    sns.heatmap(pivot_count, annot=True, fmt='.0f', cmap='Blues',
                cbar_kws={'label': 'Nombre d\'observations'}, ax=axes[1],
                linewidths=2, linecolor='white')
    axes[1].set_title(f'Heatmap: Effectifs par {cat1} et {cat2}',
                     fontsize=14, fontweight='bold', pad=15)
    axes[1].set_xlabel(cat2, fontweight='bold')
    axes[1].set_ylabel(cat1, fontweight='bold')

    plt.suptitle(f'üî• Analyse Crois√©e: {target_var} par {cat1} et {cat2}',
                 fontsize=18, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()

    # 8.4.3 Grouped bar plot
    fig, ax = plt.subplots(figsize=(12, 6))

    pivot_mean.plot(kind='bar', ax=ax, edgecolor='black', alpha=0.8, width=0.8)
    ax.set_xlabel(cat1, fontweight='bold', fontsize=12)
    ax.set_ylabel(f'{target_var} Moyen ($)', fontweight='bold', fontsize=12)
    ax.set_title(f'{target_var} Moyen par {cat1} et {cat2}',
                fontsize=14, fontweight='bold', pad=15)
    ax.legend(title=cat2, loc='best')
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

    # 8.4.4 Facet grid (si applicable)
    if df[cat2].nunique() <= 4:
        print("\nüìä Visualisation facet grid:\n")

        g = sns.FacetGrid(df, col=cat2, height=4, aspect=1.2, col_wrap=2)
        g.map_dataframe(sns.boxplot, x=cat1, y=target_var, palette='Set2')
        g.set_axis_labels(cat1, f'{target_var} ($)', fontweight='bold')
        g.set_titles(col_template="{col_name}", fontweight='bold', size=12)

        # Rotation des labels
        for ax in g.axes.flat:
            ax.grid(axis='y', alpha=0.3, linestyle='--')
            ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

        g.fig.suptitle(f'üìä Distribution de {target_var} par {cat1} (facett√© par {cat2})',
                      fontsize=16, fontweight='bold', y=1.02)
        plt.tight_layout()
        plt.show()

# 8.5 Analyse de segmentation par la variable cible
print("\n" + "-" * 80)
print(f"üîç 8.5 ANALYSE DE SEGMENTATION PAR {target_var}\n")
print("-" * 80 + "\n")

if target_var in df.columns and len(numerical_features) >= 1:

    # Cr√©er des segments bas√©s sur la variable cible
    df_segment = df.copy()
    df_segment[f'{target_var}_segment'] = pd.qcut(
        df_segment[target_var],
        q=3,
        labels=['Bas', 'Moyen', '√âlev√©'],
        duplicates='drop'
    )

    print(f"Segmentation en 3 groupes bas√©e sur {target_var}:\n")

    # Distribution des segments
    segment_counts = df_segment[f'{target_var}_segment'].value_counts().sort_index()
    print("Distribution des segments:")
    for seg, count in segment_counts.items():
        pct = (count / len(df_segment) * 100)
        print(f"   ‚Ä¢ {seg}: {count} observations ({pct:.1f}%)")
    print()

    # Profil de chaque segment
    print("üìä Profil des segments:\n")

    segment_profile = df_segment.groupby(f'{target_var}_segment')[numerical_vars].mean()
    print(segment_profile.round(2))
    print()

    # Visualisation des profils
    if len(numerical_features) >= 2:
        fig, axes = plt.subplots(1, 2, figsize=(16, 6))

        # Radar chart / parallel coordinates alternative
        segment_profile_norm = (segment_profile[numerical_features] - segment_profile[numerical_features].min()) / \
                               (segment_profile[numerical_features].max() - segment_profile[numerical_features].min())

        segment_profile_norm.T.plot(kind='bar', ax=axes[0], edgecolor='black', alpha=0.8)
        axes[0].set_xlabel('Variables', fontweight='bold', fontsize=12)
        axes[0].set_ylabel('Valeur Normalis√©e', fontweight='bold', fontsize=12)
        axes[0].set_title('Profil des Segments (Valeurs Normalis√©es)',
                         fontsize=14, fontweight='bold', pad=15)
        axes[0].legend(title='Segment')
        axes[0].grid(axis='y', alpha=0.3, linestyle='--')
        axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right')

        # Boxplot du premier feature par segment
        first_feature = numerical_features[0]
        sns.boxplot(data=df_segment, x=f'{target_var}_segment', y=first_feature,
                   ax=axes[1], palette='viridis')
        axes[1].set_xlabel(f'Segment de {target_var}', fontweight='bold', fontsize=12)
        axes[1].set_ylabel(first_feature, fontweight='bold', fontsize=12)
        axes[1].set_title(f'Distribution de {first_feature} par Segment',
                         fontsize=14, fontweight='bold', pad=15)
        axes[1].grid(axis='y', alpha=0.3, linestyle='--')

        plt.suptitle(f'üéØ Analyse de Segmentation par {target_var}',
                     fontsize=18, fontweight='bold', y=1.02)
        plt.tight_layout()
        plt.show()

    # Composition des segments par variables cat√©gorielles
    if len(categorical_vars) >= 1:
        print("\nüìä Composition des segments par variables cat√©gorielles:\n")

        for cat_var in categorical_vars[:2]:  # Limiter aux 2 premi√®res
            print(f"\n{cat_var}:")
            composition = pd.crosstab(df_segment[f'{target_var}_segment'],
                                     df_segment[cat_var],
                                     normalize='index') * 100
            print(composition.round(1))
            print()

# R√©sum√©
print("\n" + "=" * 80)
print("‚úì Section 8 termin√©e: Analyse Multivari√©e compl√®te")
print("=" * 80)
print("\nüí° POINTS CL√âS DE L'ANALYSE MULTIVARI√âE:")
print("   ‚Ä¢ Relations complexes entre plusieurs variables identifi√©es")
print("   ‚Ä¢ Patterns et clusters visuellement d√©tect√©s")
print("   ‚Ä¢ Interactions entre variables cat√©gorielles et num√©riques analys√©es")
print("   ‚Ä¢ Segmentation de la population effectu√©e")
print()
```
